```python
companyname = df['make'].apply(lambda x : x.split(' ')[0])
df.insert(2,"companyname",companyname)
df.drop(['make'],axis=1,inplace=True)
df.head()
```

```python
df['companyname'] = df['companyname'].str.lower()
df['companyname'].unique()
```

```python
df['companyname'].replace('maxda','mazda',inplace = True)
df['companyname'].replace('toyouta','toyota',inplace = True)
df['companyname'].replace('vokswagen','volkswagen',inplace = True)
df['companyname'].replace('vw','volkswagen',inplace = True)
df['companyname'].replace('porcshce','porsche',inplace = True)
```

```python
features = df.corr().sort_values(by="price", ascending=False).iloc[:11,:].index
features
```

```python
df1 = df[features]
df1
```

```python
brand = pd.get_dummies(df['fueltype'], drop_first = True)
df1 = pd.concat([df1, brand], axis = 1)
doors = pd.get_dummies(df['doornumber'], drop_first = True)
df1 = pd.concat([df1, doors], axis = 1)
carbody = pd.get_dummies(df['carbody'], drop_first = True)
df1 = pd.concat([df1, carbody], axis = 1)
df1
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df1.iloc[:,1:]
y = df1['price']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.33,random_state=1)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
plt.figure(figsize=(12,6))

sns.scatterplot(x=y_test,y=pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,pred)
```

```python
print(model.intercept_)
```

```python
cdf = pd.DataFrame(model.coef_,X.columns,columns=['Coeff'])
```
